```python
# pandas is a software library written for the Python programming language for data manipulation and analysis.
import pandas as pd
#NumPy is a library for the Python programming language, adding support for large, multi-dimensional arrays and matrices, along with a large collection of high-level mathematical functions to operate on these arrays
import numpy as np
# Matplotlib is a plotting library for python and pyplot gives us a MatLab like plotting framework. We will use this in our plotter function to plot data.
import matplotlib.pyplot as plt
#Seaborn is a Python data visualization library based on matplotlib. It provides a high-level interface for drawing attractive and informative statistical graphics
import seaborn as sns
```

```python
from js import fetch
import io

URL = "https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBM-DS0321EN-SkillsNetwork/datasets/dataset_part_2.csv"
resp = await fetch(URL)
dataset_part_2_csv = io.BytesIO((await resp.arrayBuffer()).to_py())
df=pd.read_csv(dataset_part_2_csv)
df.head(5)
```

```python
sns.catplot(y="PayloadMass", x="FlightNumber", hue="Class", data=df, aspect = 5)
plt.xlabel("Flight Number",fontsize=20)
plt.ylabel("Pay load Mass (kg)",fontsize=20)
plt.show()
```

```python
# Assuming you already have your dataframe df containing the SpaceX data

# Import necessary libraries
import seaborn as sns
import matplotlib.pyplot as plt

# Set the style for the plots
sns.set(style="whitegrid")

# Create a categorical scatter plot using catplot
plt.figure(figsize=(12, 6))
sns.catplot(x='FlightNumber', y='LaunchSite', hue='Class', data=df, kind='swarm', palette='viridis', height=6, aspect=2)

# Set plot labels and title
plt.xlabel('Flight Number')
plt.ylabel('Launch Site')
plt.title('Flight Number vs Launch Site with Class')

# Show the plot
plt.show()
```

```python
# Plot a scatter point chart with x axis as Flight Number, y axis as Launch Site, and hue as Class
plt.figure(figsize=(12, 8))
sns.scatterplot(x='FlightNumber', y='LaunchSite', hue='Class', data=df, palette='viridis', marker='o', s=100)

# Set plot labels and title
plt.xlabel('Flight Number')
plt.ylabel('Launch Site')
plt.title('Flight Number vs Launch Site with Class Hue')

# Show the plot
plt.show()
```

```python
# Plot a scatter point chart with x axis as Payload Mass (kg), y axis as Launch Site, and hue as Class
plt.figure(figsize=(12, 8))
sns.scatterplot(x='PayloadMass', y='LaunchSite', hue='Class', data=df, palette='viridis', marker='o', s=100)

# Set plot labels and title
plt.xlabel('Payload Mass (kg)')
plt.ylabel('Launch Site')
plt.title('Payload Mass vs Launch Site with Class Hue')

# Show the plot
plt.show()
```

```python
# Plot a bar chart for the success rate of each orbit
plt.figure(figsize=(12, 8))
orbit_success_rate = df.groupby('Orbit')['Class'].mean().sort_values(ascending=False)
sns.barplot(x=orbit_success_rate.index, y=orbit_success_rate.values, palette='viridis')

# Set plot labels and title
plt.xlabel('Orbit Type')
plt.ylabel('Success Rate')
plt.title('Success Rate of Each Orbit Type')

# Show the plot
plt.show()
```

```python
# Plot a scatter point chart with x axis as FlightNumber, y axis as Orbit, and hue as Class
plt.figure(figsize=(12, 8))
sns.scatterplot(x='FlightNumber', y='Orbit', hue='Class', data=df, palette='viridis', marker='o', s=100)

# Set plot labels and title
plt.xlabel('Flight Number')
plt.ylabel('Orbit Type')
plt.title('Flight Number vs Orbit Type with Class Hue')

# Show the plot
plt.show()
```

```python
# Plot a scatter point chart with x axis as Payload, y axis as Orbit, and hue as Class
plt.figure(figsize=(12, 8))
sns.scatterplot(x='PayloadMass', y='Orbit', hue='Class', data=df, palette='viridis', marker='o', s=100)

# Set plot labels and title
plt.xlabel('Payload Mass (kg)')
plt.ylabel('Orbit Type')
plt.title('Payload Mass vs Orbit Type with Class Hue')

# Show the plot
plt.show()
```

```python
# Extract the year from the Date column
df['Year'] = pd.to_datetime(df['Date']).dt.year

# Plot a line chart with x axis as Year and y axis as average success rate
plt.figure(figsize=(12, 8))
yearly_success_rate = df.groupby('Year')['Class'].mean()
sns.lineplot(x=yearly_success_rate.index, y=yearly_success_rate.values, marker='o', color='blue')

# Set plot labels and title
plt.xlabel('Year')
plt.ylabel('Average Success Rate')
plt.title('Launch Success Yearly Trend')

# Show the plot
plt.show()
```

```python
# A function to Extract years from the date 
year=[]
def Extract_year():
    for i in df["Date"]:
        year.append(i.split("-")[0])
    return year
Extract_year()
df['Date'] = year
df.head()
    
```

```python
# A function to Extract years from the date
def Extract_year(date):
    return date.split("-")[0]

# Apply the function to create a new 'Year' column
df['Year'] = df['Date'].apply(Extract_year)

# Plot a line chart with x axis as extracted Year and y axis as average success rate
plt.figure(figsize=(12, 8))
yearly_success_rate = df.groupby('Year')['Class'].mean()
sns.lineplot(x=yearly_success_rate.index, y=yearly_success_rate.values, marker='o', color='green')

# Set plot labels and title
plt.xlabel('Year')
plt.ylabel('Average Success Rate')
plt.title('Launch Success Yearly Trend')

# Show the plot
plt.show()
```

```python
features = df[['FlightNumber', 'PayloadMass', 'Orbit', 'LaunchSite', 'Flights', 'GridFins', 'Reused', 'Legs', 'LandingPad', 'Block', 'ReusedCount', 'Serial']]
features.head()
```

```python
# Applying OneHotEncoder using get_dummies on categorical columns
features_one_hot = pd.get_dummies(features, columns=['Orbit', 'LaunchSite', 'LandingPad', 'Serial'])

# Display the head of the DataFrame with one-hot encoded features
features_one_hot.head()
```

```python
# Cast the entire dataframe to float64
features_one_hot = features_one_hot.astype('float64')
```

```python
features_one_hot.to_csv('dataset_part_3.csv', index=False)
```
